# 1. Import dependencies

```python
import pandas as pd
```

# 2. Load data

```python
data = pd.read_csv("/home/rakesh47/NNfromScratch/Data/boston_housing_data.csv", delim_whitespace=1, header=None)
data.head()
```

# 3. Prepare data

## 3.1. Create train and test sets

### 3.1.1. Randomly shuffle data

```python
data = data.sample(frac=1).reset_index(drop=True)
data.head()
```

### 3.1.2. Create train and test sets with 70:30 ratio

```python
n_train, n_test = int(0.7 * len(data)), int(0.3 * len(data))
n_train, n_test
```

```python
train, test = data.iloc[:n_train, :], data.iloc[n_train:, :]
```

```python
train.head()
```

```python
test.head()
```

## 3.2. Create X_train, y_train, X_test, y_test sets

### 3.2.1. Create X_train, y_train sets

```python
X_train, y_train = train.iloc[:, 0:-1], train.iloc[:, -1]
```

```python
X_train.head()
```

```python
y_train.head()
```

### 3.2.2. Create X_train, y_train sets

```python
X_test, y_test = test.iloc[:, 0:-1], test.iloc[:, -1]
```

```python
X_test.head()
```

```python
y_test.head()
```

# 4. Save prepared data

```python
X_train.to_csv("/home/rakesh47/NNfromScratch/PreparedData/X_train.csv", index=None, header=None)
y_train.to_csv("/home/rakesh47/NNfromScratch/PreparedData/y_train.csv", index=None, header=None)

X_test.to_csv("/home/rakesh47/NNfromScratch/PreparedData/X_test.csv", index=None, header=None)
y_test.to_csv("/home/rakesh47/NNfromScratch/PreparedData/y_test.csv", index=None, header=None)
```
